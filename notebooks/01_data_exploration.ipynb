```python
df.columns = (
    df.columns
    .str.lower()
    .str.strip()
    .str.replace(" ", "_")
)

df.columns
```

```python
df = df.drop_duplicates()

df.shape
```

```python
df.head()
```

```python
df = df.drop(
    columns=[
        "description",
        "closed_datetime",
        "action_taken_timestamp"
    ]
)

df.shape
```

```python
df["location"] = df["location"].fillna("Unknown")
```

```python
df["police_station"] = (
    df["police_station"]
    .fillna("Unknown")
)
```

```python
df["junction_name"] = (
    df["junction_name"]
    .fillna("Unknown")
)
```

```python
df.isnull().sum()
```

```python
df["created_datetime"] = pd.to_datetime(
    df["created_datetime"],
    format="mixed",
    utc=True
)
```

```python
df["created_datetime"].head()
```

```python
df.info()
```

```python
#Extracting the hr in the date
df["hour"] = df["created_datetime"].dt.hour
```

```python
df[["created_datetime","hour"]].head()
```

```python
#Day of the week
df["day_of_week"] = (
    df["created_datetime"]
    .dt.day_name()
)
```

```python
df["day_of_week"].value_counts()
```

```python
#Month
df["month"] = (
    df["created_datetime"]
    .dt.month
)
```

```python
#Weekend
df["is_weekend"] = (
    df["created_datetime"]
    .dt.dayofweek >= 5
)
```

```python
#Understanding peak hours
def get_time_period(hour):

    if 6 <= hour < 12:
        return "Morning"

    elif 12 <= hour < 17:
        return "Afternoon"

    elif 17 <= hour < 21:
        return "Evening Peak"

    else:
        return "Night"


df["time_period"] = (
    df["hour"]
    .apply(get_time_period)
)
```

```python
df["time_period"].value_counts()
```

```python
#Vehicle type distribution
df["vehicle_type"].value_counts()
```

```python
#Vehicle impact score
vehicle_weights = {

    # two wheelers - low blockage
    "SCOOTER": 1,
    "MOTOR CYCLE": 1,
    "MOPED": 1,


    # medium vehicles
    "CAR": 3,
    "PASSENGER AUTO": 3,
    "GOODS AUTO": 3,
    "JEEP": 3,
    "VAN": 3,


    # commercial medium-heavy
    "MAXI-CAB": 4,
    "LGV": 4,
    "TEMPO": 4,


    # heavy vehicles
    "PRIVATE BUS": 5,
    "BUS (BMTC/KSRTC)": 5,
    "TOURIST BUS": 5,
    "SCHOOL VEHICLE": 5,
    "FACTORY BUS": 5,
    "HGV": 5,
    "LORRY/GOODS VEHICLE": 5,
    "TANKER": 5,
    "MINI LORRY": 5,
    "TRACTOR": 5,

    "OTHERS": 2
}


df["vehicle_impact"] = (
    df["vehicle_type"]
    .map(vehicle_weights)
)
```

```python
df[["vehicle_type","vehicle_impact"]].head(20)
```

```python
df["vehicle_impact"].isnull().sum()
```

```python
df["violation_type"].value_counts()
```

```python
#Adding violation count
df["violation_count"] = (
    df["violation_type"]
    .str.count(",") + 1
)
```

```python
df[
    ["violation_type", "violation_count"]
].head(20)
```

```python
# Road obstruction impact caused by violation type
def calculate_obstruction_score(x):

    score = 1

    if "WRONG PARKING" in x:
        score += 1

    if "NO PARKING" in x:
        score += 2

    if "PARKING IN A MAIN ROAD" in x:
        score += 5

    if "PARKING NEAR ROAD CROSSING" in x:
        score += 6

    if "PARKING NEAR TRAFFIC LIGHT" in x:
        score += 6

    if "PARKING NEAR BUSTOP" in x:
        score += 5

    if "DOUBLE PARKING" in x:
        score += 7

    return score


df["obstruction_score"] = (
    df["violation_type"]
    .apply(calculate_obstruction_score)
)
```

```python
df[
[
"violation_type",
"obstruction_score"
]
].sort_values(
    "obstruction_score",
    ascending=False
).head(10)
```

```python
#Time impact
def calculate_time_impact(hour):

    if 8 <= hour <= 10:
        return 5

    elif 17 <= hour <= 20:
        return 5

    elif 11 <= hour <= 16:
        return 3

    else:
        return 1


df["time_impact"] = (
    df["hour"]
    .apply(calculate_time_impact)
)
```

```python
df["enforcement_zone"] = df["junction_name"]

df.loc[
    df["enforcement_zone"] == "No Junction",
    "enforcement_zone"
] = df["location"]
```

```python
df["enforcement_zone"].value_counts().head(10)
```

```python
df = df.drop(
    columns=["location_pressure"],
    errors="ignore"
)
```

```python
#Calculating location pressure
zone_frequency = (
    df["enforcement_zone"]
    .value_counts()
)


df["location_pressure"] = (
    df["enforcement_zone"]
    .map(zone_frequency)
)


df["location_pressure"] = (

    df["location_pressure"]  / df["location_pressure"].max() * 10
)
```

```python
df["congestion_impact_score"] = (

    df["obstruction_score"] * 0.35

    +

    df["vehicle_impact"] * 0.25

    +

    df["time_impact"] * 0.20

    +

    df["location_pressure"] * 0.20
)


df["congestion_impact_score"] = (

    df["congestion_impact_score"]

    /

    df["congestion_impact_score"].max()

    *

    100

).round(2)
```

```python
df[
[
"enforcement_zone",
"vehicle_type",
"hour",
"congestion_impact_score"
]
].sort_values(
    "congestion_impact_score",
    ascending=False
).head(15)
```

```python
df.to_csv(
    "../data/processed/cleaned_parking_data.csv",
    index=False
)
```
